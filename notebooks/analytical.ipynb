```python
import matplotlib.pyplot as plt
import numpy as np
from math import exp
```

```python
i0 = 0.001
b = 1.2
d = 1/5
r = b/d

i = lambda t: i0 * (b-d)   /   ( i0 * b * (1 - exp(-(b-d)*t))  + (b-d) * exp(-(b-d)*t) )


t_list = np.linspace(0,20,100)


i_list = np.array([i(t) for t in t_list])
s_list = np.array([1-i  for i in i_list])
r_list = np.array([r*s  for s in s_list])


fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,3), dpi=180)
fig.suptitle('SIS')


ax1.plot(t_list, s_list, label='Susceptibles')
ax1.plot(t_list, i_list, label='Infectives')
ax1.legend()

ax2.plot(t_list, r_list, label='effective rate of reproduction', color='violet')
ax2.legend()

plt.show()
```
