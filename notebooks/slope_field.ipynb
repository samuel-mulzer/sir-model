```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def phase_plane_portrait(xmin, xmax, ymin, ymax, length):
    # draw phase plane portrait with maptlotlib quiver
    x = np.arange(xmin, xmax, length)
    y = np.arange(ymin, ymax, length)
    
    X, Y = np.meshgrid(x, y)

    U = 1
    V = 2 * Y

    U = length * U / np.sqrt(U**2 + V**2)
    V = length * V / np.sqrt(U**2 + V**2)

    plt.quiver(X, Y, U, V, units='xy', angles='xy', scale_units='xy')
    plt.show()

phase_plane_portrait(0, 10, 0, 2, 0.5)
```
